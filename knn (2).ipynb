```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from collections import Counter
import numpy as np
```

```python
df = pd.read_csv("Our Dataset.csv")
```

```python
df.head()
```

```python
dataset = {}

for sport in df['Sports'].unique():
    players = df[df['Sports'] == sport][['height','weight']].values
    dataset[sport]=players.tolist()

print(dataset)
```

```python
def scatter_plot(data):
    for sports  in data:
        if sports == 'Cricket':
            color = 'r'
        else:
            color = 'b'
        for player in data[sports]:
            plt.scatter(player[0],player[1],label=sports,color=color)
    plt.show()
```

```python
scatter_plot(dataset)
```

```python
def scatter_plot(data, new_player):
    for sports in data:
        if sports == 'Cricket':
            color = 'r'
        else:
            color = 'g'

        for player in data[sports]:
            plt.scatter(player[0], player[1],
                        label=sports,
                        color=color)

    plt.scatter(new_player[0], new_player[1],
                label='New_player',
                color='black',
                s=100)

    plt.show()
```

```python
def K_NN(data,new_player,k):
    distances=[]
    for sports in data:
        for player in data[sports]:
            distance = np.sqrt((player[0] - new_player[0])**2+(player[1] - new_player[1])**2)
            distances.append([distance,sports])

    votes=[vote[1] for vote in sorted(distances)[:k]]
    vote_result = Counter(votes).most_common()[0][0]
    return vote_result

```

```python
print(K_NN(dataset,[182,60],5))



```
