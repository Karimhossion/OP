```python
from collections import deque

graph = {
    "S": ["I", "A", "L", "B", "J"],
    "I": ["C"],
    "A": ["I", "C", "D", "L"],
    "L": [],
    "B": ["L", "E", "F", "J"],
    "J": ["F"],
    "C": ["H"],
    "D": ["H", "E"],
    "E": ["G", "F"],
    "F": [],
    "H": [],
    "G": []
}
```

```python
def bfs_search(graph, start_node, goal_node):
    visited = set()
    queue = deque([start_node])
    bfs_order = []

    while queue:
        current_node = queue.popleft()

        if current_node not in visited:
            visited.add(current_node)
            bfs_order.append(current_node)

            if current_node == goal_node:
                break

            for next_node in graph[current_node]:
                if next_node not in visited:
                    queue.append(next_node)

    return bfs_order

result = bfs_search(graph, "S", "G")
print(result)
```

```python
def dfs_search(graph, start_node, goal_node):
    visited = set()
    stack = [start_node]      # DFS uses Stack (LIFO)
    dfs_order = []

    while stack:
        current_node = stack.pop()

        if current_node not in visited:
            visited.add(current_node)
            dfs_order.append(current_node)

            if current_node == goal_node:
                break

            # Reverse order for correct DFS traversal
            for next_node in reversed(graph[current_node]):
                if next_node not in visited:
                    stack.append(next_node)

    return dfs_order

result = dfs_search(graph, "S", "G")
print(result)
```
